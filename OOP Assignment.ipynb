```python
class Employee:
    def __init__(self,name,number,address,salary,post,year_of_joining,education,year_of_experience):
        self.name=name
        self.number=number
        self.address=address
        self.salary=salary
        self.post=post
        self.year_of_joining=year_of_joining
        self.education=education
        self.year_of_experience=year_of_experience

    def describe_employee(self):
        print("\nDetailed information of Employee:")
        print(f"Employee name is: {self.name}")
        print(f"Employee post is: {self.post}")
        print(f"Employee education is: {self.education}")
        print(f"Employee salary is: {self.salary}")
        print(f"Employee number is: {self.number}")
        print(f"Employee address is: {self.address}")
        print(f"Employee year of joining is: {self.year_of_joining}")
        print(f"Employee year of experience is: {self.year_of_experience}")
        
        

    def update_year_of_experience(self,add_update):
        self.year_of_experience += add_update

    def increase_salary(self, increament):
        self.salary += increament

employee101=Employee("Shahmir","03128333831","Pakistan",99999,"Data Engineer","2011","Masters in DS",10)
print(employee101.name)
print(employee101.salary)
print(employee101.post)
print(employee101.education)

#method describe_employee()
# employee101.describe_employee()

#method increase_salary()
employee101.increase_salary(119191)

#method update_ year_of_experience()
employee101.update_year_of_experience(6)


employee101.describe_employee()

#different instances from the class
employee102=Employee("Ali","03217266261","Australia",33782,"Graphics Designer","2009","Masters",1)
employee103=Employee("Haseeb","03461827227","USA",88292,"Data Scientist","2005","Phd",2)
employee104=Employee("Tahir","044488837373","UK",882920,"Software Engineer","2023","BS",4)
employee105=Employee("Uzma","092727552512","Germany",222902,"Web developer","2019","M.phill",10)
employee106=Employee("Azam","099827117111","Malaysia",8827171,"App developer","2011","Diploma",5)

#call and update employee102
employee102.increase_salary(6666525)
employee102.update_year_of_experience(1)
employee102.describe_employee()

#call and update employee103
employee103.increase_salary(28282)
employee103.update_year_of_experience(2)
employee103.describe_employee()

#call and update employee104
employee104.increase_salary(9818)
employee104.update_year_of_experience(4)
employee104.describe_employee()

#call and update employee105
employee105.increase_salary(7772)
employee105.update_year_of_experience(7)
employee105.describe_employee()

```
