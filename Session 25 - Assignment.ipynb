### ***1. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000***

```python
q = '''select jobtitle,overtimepay from salaries
        where overtimepay > 50000
        and
        overtimepay is not null
        and
        overtimepay != 'Not Provided'
'''
sql_query(q)
```

## ***2. Show the Average BasePay for all Employees.***

### ***4. Show The Average of BasePay, OverTime, OtherPay for Each Employee***

```python
q = '''select Employeename, (basepay + overtimepay + otherpay)/3 as Avgerage from salaries
    where Employeename is not null
    and 
    employeename != 'Not provided'

'''
sql_query(q)
```

### ***5. Show All Employees Who Have The Word "Manager" In Their Job Title***

```python
q = '''select employeename,jobtitle from salaries where jobtitle like '%manager%'
'''
sql_query(q)
```

### ***6. Show All Employees With A Job Title Not Equal to "Manager"***

```python
q = '''select employeename,jobtitle from salaries where jobtitle not like '%manager%'
        and jobtitle != 'Not provided'
'''
sql_query(q)
```

### ***7. Show All Employess With A Total Pay Between 50,000 and 75,000***

```python
q = '''select employeename,totalpay from salaries
    where totalpay between 50000 and 75000'''
sql_query(q)
```

### ***8. Show All Employees With A BasePay less than 50,000 Or A TotalPay Greater Than 100,000***

```python
q = '''select EmployeeName,basepay from salaries
    where BasePay < 50000 or TotalPay > 100000
    '''

sql_query(q)
```

### ***9. Show All Employees With A TotalPay Benefits Value Between 125,000 and 150,000 And A Job Title Containing Word 'Director'***

```python
q = '''select Employeename,jobtitle from salaries
    where TotalPay between 125000 and 150000 and JobTitle like '%Director%';
'''

sql_query(q)
```

### ***10. Show All Employees Ordered By Their TotalPay Benefits In Descending Order***

```python
q = '''select employeename,totalpaybenefits from salaries
    order by totalpaybenefits desc
'''

sql_query(q)
```

### ***11. How many employees in the salaries table have OvertimePay equal to 0?***

```python
q = '''select count(*) as EmployeeID,employeename,overtimepay from salaries
    where OvertimePay = 0
'''

sql_query(q)
```

## ***12. Display Name of The Person Having The Highest BasePay***

```python
q = '''select max(basepay) as 'maximum basepay',employeename from salaries
    where basepay != 'not provided' and basepay GLOB '[0-9]*'
    '''

sql_query(q)
```

## ***13. Find Top 5 Most Common Jobs***

```python
q = '''select jobtitle, COUNT(*) as job_count from salaries
    group by jobtitle
    order by job_count desc
'''
sql_query(q)
```

## ***14. Display Name of The Person Having The Lowest TotalPay***

```python
q = '''select min(totalpay) as 'minimum totalpay',employeename from salaries
    where totalpay != 'not provided' and totalpay GLOB '[0-9]*'
    '''

sql_query(q)
```

## ***15. Number of Job Titles repeatd only once***

```python
q = '''SELECT COUNT(*) AS Unique_Job_Titles
from (
    select
        jobtitle
    from
        salaries
    group by 
        jobtitle
    having
        count(*) = 1
)'''

sql_query(q)
```
