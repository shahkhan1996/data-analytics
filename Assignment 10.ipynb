## ***13. Find Total Number of Persons Having Age Between 17 To 48 (Inclusive) Using Between Method And Also Without Using Between Method***

```python
count_between = df['age'].between(17, 48).sum()

count_without_between = ((df['age'] >= 17) & (df['age'] <= 48)).sum()

print(f"Total number of persons with age between 17 and 48 (using between): {count_between}")
print(f"Total number of persons with age between 17 and 48 (without using between): {count_without_between}")
```

## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})
df['income'] = df['income'].astype(int)
df['income']
```

## ***18. Which Workclass getting The Highest Salary***

```python
workclass_income_counts = df.groupby(['workclass', 'income']).size().unstack(fill_value=0)

print("Count of each workclass category by income:")
print(workclass_income_counts)

highest_income_workclass = workclass_income_counts[1].idxmax()
highest_income_count = workclass_income_counts[1].max()

print(f"\nThe workclass with the highest number of individuals income is: {highest_income_workclass} with {highest_income_count} individuals.")
```

```python
workclass_income_counts = df.groupby(['workclass', 'income']).size().unstack(fill_value=0)

workclass_income_counts.columns = ['<=50K', '>50K']

print("Count of each workclass category by income:")
print(workclass_income_counts)

workclass_income_counts['Total'] = workclass_income_counts.sum(axis=1)
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
gender_counts = df.groupby('gender')['income'].value_counts().unstack(fill_value=0)

gender_counts['Proportion >50K'] = gender_counts[1] / (gender_counts[0] + gender_counts[1])

print("Counts and proportions of income >50K by gender:")
print(gender_counts[[1, 0, 'Proportion >50K']])

better_chance_gender = gender_counts['Proportion >50K'].idxmax()
better_chance_value = gender_counts['Proportion >50K'].max()

print(f"\nThe gender with a better chance of earning >50K is: {better_chance_gender} with a proportion of {better_chance_value:2f}.")
```

## ***21. Graphical representaion of Education level by gender***

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='education', hue='gender', palette='colorblind')

plt.title('Count of Education Levels by Gender', fontsize=16)
plt.xlabel('Education Level', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## ***22. Heatmap of Race vs Workclass***

```python
counts = df.groupby(['race', 'workclass'], observed=False).size().reset_index(name='count')

heatmap_data = counts.pivot(index='race', columns='workclass', values='count').fillna(0)

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='magma', fmt='.0f')

plt.title('Heatmap of Race vs. Workclass', fontsize=16)
plt.xlabel('Workclass', fontsize=14)
plt.ylabel('Race', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
