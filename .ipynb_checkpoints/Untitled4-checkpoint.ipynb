```python
class Employee:
    def __init__(self, employee_dict):
        self.employee = employee_dict

    def add_employee(self, id, name, salary):
        if id in self.employee:
            raise ValueError(f"Employee with {id} already exists")
        self.employee[id] = {
            "Employee ID" : id,
            "Name" : name,
            "Salary" : salary
        }

    def delete_employee(self, id):
        if id not in self.employee:
            raise ValueError(f"Employee with {id} does not exist")
        del self.employee[id]

    def display_employee(self, id):
        if id not in self.employee:
            raise ValueError (f"Employee with {id} does not exist")
        return self.employee[id]

    def list_employee(self):
        return list(self.employee.values())

    def update_employee(self, id, name=None, salary=None):
        if id not in self.employee:
            raise ValueError (f"Employee with {id} does not exist")
        if name is not None:
            self.employee[id]["Name"] = name
        if salary is not None:
            self.employee[id]["Salary"] = salary


def main():

    employee_data = {
        101: {'Employee ID': 101, 'Name': 'Alice Smith', 'Position': 'Software Engineer', 'Salary': 85000},
        102: {'Employee ID': 102, 'Name': 'Bob Jones', 'Position': 'Data Scientist', 'Salary': 95000}
    }
    
    record = Employee(employee_data)

    while True:
        print("\nEmployee Record Management System")
        print("1. Add Employee")
        print("2. Update Employee")
        print("3. Delete Employee")
        print("4. Display Employee")
        print("5. List All Employees")
        print("6. Exit")

        choice = input("Enter your choice (1-6)")

        if choice == "1":
            try:
                id = int(input("Enter Employee ID: "))
                name = input("Enter Employee Name: ")
                salary = float(input("Enter Employee Salary: "))
                record.add_employee(id, name, salary)
                print("Employee Record added")
            except ValueError as e:
                print(f"Error: {e}")
                
        elif choice == "2":
            try:
                id = int(input("Enter Employee ID: "))
                name = input("Enter Employee Name (leave empty to keep current): ")
                salary_input = input("Enter Employee Salary (leave empty to keep current): ")
                salary = float(salary_input) if salary_input else None
                record.update_employee(id, name if name else None, salary if salary else None)
                print("Employee updated")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "3":
            try:
                id = int(input("Enter Employee ID to delete: "))
                record.delete_employee(id)
                print("Employee deleted")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "4":
            try:
                id = int(input("Enter Employee ID to display: "))
                employee = record.display_employee(id)
                print("Employee Details: ", employee)
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "5":
            employees = record.list_employee()
            if employees:
                print("All Employee list: ")
                for emp in employees:
                    print(emp)
            else:
                print("No Employee found")
            
        elif choice == "6":
            print("Exit")
            break

        else:
            print("Invalid choice. please enter 1-6")


main()

            
                
                
                
                
```

```python
# correction by python

class Employee:
    def __init__(self, employee_dict):
        self.employee = employee_dict

    def add_employee(self, id, name, salary, position):
        if id in self.employee:
            raise ValueError(f"Employee with ID {id} already exists")
        self.employee[id] = {
            "Employee ID": id,
            "Name": name,
            "Position": position,
            "Salary": salary
        }

    def update_employee(self, id, name=None, position=None, salary=None):
        if id not in self.employee:
            raise ValueError(f"Employee with ID {id} does not exist")
        if name is not None:
            self.employee[id]["Name"] = name
        if position is not None:
            self.employee[id]["Position"] = position
        if salary is not None:
            self.employee[id]["Salary"] = salary

    def delete_employee(self, id):
        if id not in self.employee:
            raise ValueError(f"Employee with ID {id} does not exist")
        del self.employee[id]

    def display_employee(self, id):
        if id not in self.employee:
            raise ValueError(f"Employee with ID {id} does not exist")
        return self.employee[id]

    def list_employee(self):
        return list(self.employee.values())

def main():
    employee_data = {
        101: {'Employee ID': 101, 'Name': 'Alice Smith', 'Position': 'Software Engineer', 'Salary': 85000},
        102: {'Employee ID': 102, 'Name': 'Bob Jones', 'Position': 'Data Scientist', 'Salary': 95000}
    }

    record = Employee(employee_data)

    while True:
        print("\nEmployee Record Management Application")
        print("1. Add Employee")
        print("2. Update Employee")
        print("3. Delete Employee")
        print("4. Display Employee")
        print("5. List All Employees")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            try:
                id = int(input("Enter Employee ID: "))
                name = input("Enter Employee Name: ")
                position = input("Enter Employee Position: ")
                salary = float(input("Enter Employee Salary: "))
                record.add_employee(id, name, salary, position)
                print("Employee added successfully")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == '2':
            try:
                id = int(input("Enter Employee ID: "))
                name = input("Enter New Employee Name (leave empty to keep current): ")
                position = input("Enter New Employee Position (leave empty to keep current): ")
                salary_input = input("Enter New Employee Salary (leave empty to keep current): ")
                salary = float(salary_input) if salary_input else None
                record.update_employee(id, name if name else None, position if position else None, salary)
                print("Employee updated successfully")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == '3':
            try:
                id = int(input("Enter Employee ID to delete: "))
                record.delete_employee(id)
                print("Employee deleted successfully.")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == '4':
            try:
                id = int(input("Enter Employee ID to display record: "))
                employee = record.display_employee(id)
                print("Employee Details:", employee)
            except ValueError as e:
                print(f"Error: {e}") 

        elif choice == '5':
            employees = record.list_employee()
            if employees:
                print("All Employees List:")
                for emp in employees:
                    print(emp)
            else:
                print("No Employees found")

        elif choice == '6':
            print("Exiting application")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```
