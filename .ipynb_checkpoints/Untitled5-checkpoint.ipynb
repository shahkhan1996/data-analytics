```python
class Employee:
    def __init__(self, employee_dict):
        self.employee = employee_dict

    def add_employee(self, id, name, salary):
        if id in self.employee:
            raise ValueError(f"Employee with {id} already exist")
        self.employee[id] = {
            "Employee ID" : id,
            "Name" : name,
            "Salary" : salary
        }

    def update_employee(self, id, name=None, salary=None):
        if id not in self.employee:
            raise ValueError(f"Employee with {id} does not exist")
        if name is not None:
            self.employee[id]["Name"]=name
        if salary is not None:
            self.employee[id]["Salary"] = salary

    def delete_employee(self,id):
        if id not in self.employee:
            raise ValueError(f"Employee with {id} does not exist")
        del self.employee[id]

    def display_employee(self,id):
        if id not in self.employee:
            raise ValueError(f"Employee with {id} does not exist")
        return self.employee[id]

    def list_employee(self):
        return list(self.employee.values())

def main():

    employee_data = {
        101: {"Employee ID" : 101, "Name":"Shahmir", "Salary" : 99999},
        102: {"Employee ID" : 102, "Name":"Ali", "Salary" : 7777},
    }
    
    record = Employee(employee_data)

    while True:
        print("\nEmployee Record Management System")
        print("1. Add Employee")
        print("2. Update Employee")
        print("3. Delete Employee")
        print("4. Display Employee")
        print("5. List All Employees")
        print("6. Exit")

        choice = input("Enter your choice (1-6)")

        if choice == "1":
            try:
                id = int(input("Enter Id: "))
                name = input("Enter name: ")
                salary = float(input("enter salary: "))
                record.add_employee(id,name,salary)
                print("employee added")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "2":
            try:
                id = int(input("Enter Id: "))
                name = input("Enter name (leave empty to keep current): ")
                salary = float(input("enter salary (leave empty to keep current)"))
                record.update_employee(id, name if name else None, salary if salary else None)
                print("Employee updated")
            except ValueError as e:
                print(f"Error: {e}") 

        elif choice == "3":
            try:
                id = int(input("Enter Id: "))
                record.delete_employee(id)
                print("employee deleted")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "4":
            try:
                id = int(input("Enter Id: "))
                employee = record.display_employee(id)
                print("employee details: " , employee)
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "5":
            employees = record.list_employee()
            if employees:
                print("all employees list: ")
                for emp in employees:
                    print(emp)
            else:
                print("no employees found")

        elif choice == "6":
            print("exit")
            break

        else:
            print("invalid choice")

main()
            
                
                
```
